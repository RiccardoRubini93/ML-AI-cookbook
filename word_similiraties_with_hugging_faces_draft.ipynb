```python
from sentence_transformers import SentenceTransformer, util
```

```python
sentences = ["I'm happy", "I'm full of happiness"]
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
#check similarity

sentences_ = ["That is a happy person", "That is a very happy person"]

embedding_1= model.encode(sentences_[0], convert_to_tensor=True)
embedding_2 = model.encode(sentences_[1], convert_to_tensor=True)

res = util.pytorch_cos_sim(embedding_1, embedding_2)
```

```python
#Our sentences we like to encode
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.',
    'The quick brown fox jumps over the lazy happy dog.',
            'I m happy', 'happiness']

#Sentences are encoded by calling model.encode()
embeddings = model.encode(sentences)

#Print the embeddings
for sentence, embedding in zip(sentences, embeddings):
    print("Sentence:", sentence)
    #print("Embedding:", embedding)
    print("")
    print(f"Lenght : {len(embedding)}")
```
